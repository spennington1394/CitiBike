```python
## read in files
files = ['Resources/202001-citibike-tripdata.csv',
        'Resources/202002-citibike-tripdata.csv',
         'Resources/202003-citibike-tripdata.csv',
        'Resources/202004-citibike-tripdata.csv',
         'Resources/202005-citibike-tripdata.csv',
         'Resources/202006-citibike-tripdata.csv',
         'Resources/202007-citibike-tripdata.csv',
         'Resources/202008-citibike-tripdata.csv',
         'Resources/202009-citibike-tripdata.csv',
         'Resources/202010-citibike-tripdata.csv',
         'Resources/202011-citibike-tripdata.csv',
         'Resources/202012-citibike-tripdata.csv'
#          'Resources/202101-citibike-tripdata.csv'
        ]
## establish empty lists
user_type = []
birthdate = []
gender = []
bikeid = []
start_station = []
end_station = []
# lat = []
# long = []

## set up for loop 
for file in files:
    data = pd.read_csv(file, index_col = None)
    user_type.append(data['usertype'].value_counts())
    birthdate.append(data['birth year'].value_counts())
    gender.append(data['gender'].value_counts())
    bikeid.append(data['bikeid'].value_counts())
    start_station.append(data['start station name'].value_counts())
    end_station.append(data['end station name'].value_counts())

    print("calculating....", file)
```
